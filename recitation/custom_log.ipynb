```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
```

```python
n = np.linspace(1,100,100)
```

```python
# n
```

```python
# custom base log in numpy
base = 3

lg_n_pow2 = np.log10(n)
custom_log2 = np.log(n) / np.log(base)
custom_log2 += 2
# exponent = np.log(array) / np.log(base)
# lg_n_pow2
```

```python
# plt.plot(n,lg_n_pow2, c='y')
plt.plot(n,lg_n_pow2, c='b')
plt.plot(n,custom_log2, c='r')
# plt.plot(n,sqrt_n, c='g')
# plt.title("Matplotlib PLot NumPy Array")
# plt.xlabel("x axis")
# plt.ylabel("y axis")
plt.show()
```

### Comparing two logs

```python
log2 = np.log2(n)
log_10_div5 = np.log10(n) * 3.5
```

```python
# plt.plot(n,lg_n_pow2, c='y')
plt.plot(n,log2, c='b')
plt.plot(n,log_10_div5, c='r')
# plt.plot(n,sqrt_n, c='g')
# plt.title("Matplotlib PLot NumPy Array")
# plt.xlabel("x axis")
# plt.ylabel("y axis")
plt.show()
```
